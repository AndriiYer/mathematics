## Аналітичний розв'язок:

Маємо функцію $ f(t) = 1000 \cdot t \cdot e^{-0.2t}. $

Позначимо $ u(t) = 1000t,\quad v(t) = e^{-0.2t}. $ Тоді $ u'(t) = 1000,\quad v'(t) = -0.2 e^{-0.2t}. $

За правилом добутку $ (f(t))' = (u \cdot v)' = u'(t)v(t) + u(t)v'(t). $

Отже $ f'(t) = 1000 \cdot e^{-0.2t} + 1000t \cdot \left(-0.2 e^{-0.2t}\right) = 1000 e^{-0.2t} - 200t e^{-0.2t} = 1000 e^{-0.2t}(1 - 0.2t). $

### Аналітична похідна: $ \boxed{f'(t) = 1000 e^{-0.2t}(1 - 0.2t)}. $

### Пікове навантаження

Знайдемо максимум, розв’язавши рівняння $ f'(t) = 0. $

Оскільки $ (1000 e^{-0.2t} > 0) $ для всіх (t), маємо $ 1 - 0.2t = 0 \quad \Rightarrow \quad 0.2t = 1 \quad \Rightarrow \quad t = 5. $

Початок робочого дня: 8:00.Через (t = 5) годин:

$ 8{:}00 + 5\text{ годин} = 13{:}00. $

Момент пікового навантаження: $ \boxed{t^* = 5 \text{ годин} \Rightarrow 13{:}00.} $

### Програмний розв'язок:

```python
# вихідна функція
def f(t):
    return 1000.0 * t * np.exp(-0.2 * t)

# аналітична похідна
def f_prime_analytic(t):
    return 1000.0 * np.exp(-0.2 * t) * (1.0 - 0.2 * t)

# чисельна похідна через approx_fprime
def f_prime_numeric(t, eps=1e-6):
    # approx_fprime очікує вектор, тому обгортаємо t в np.array
    xk = np.array($t$, dtype=float)
    def f_wrapper(x):
        # x тут масив, беремо перший елемент
        return f(x$0$)
    grad = approx_fprime(xk, f_wrapper, eps)
    return grad$0$

times = $2.0, 6.0, 10.0]  # години від початку дня

print("t (години)\tаналітична f'(t)\tчисельна f'(t)")
for t in times:
    fa = f_prime_analytic(t)
    fn = f_prime_numeric(t)
    print(f"{t:>8.1f}\t{fa:>18.6f}\t{fn:>18.6f}")
```

# Задача 2. Моделювання процесу навчання

Цю залежність описує наступне диференціальне рівняння:

$\cfrac{dK}{dt} = r(M - K)$

де:

- $K(t)$ — рівень знань студента в момент часу $t$ (у відсотках, від 0 до 100)
- $M=100$ — максимальний рівень знань (100%)
- $r = 0.15$ — коефіцієнт швидкості навчання (1/день)
- Початкова умова: $K(0)=10$ (студент починає з 10% базових знань)

Наше рівняння каже нам, що чим більше студент вже знає, тим повільніше він вчить нове.

Використовуючи чисельні методи розв'язання диференціальних рівнянь, напишіть програму та виконайте наступні кроки:

Розв'язок:

```python
# ============================================================================
# 1. РЕАЛІЗАЦІЯ МОДЕЛІ
# ============================================================================

def learning_rate(t, K):
    """
    Функція, що описує праву частину диференціального рівняння навчання.

    Диференціальне рівняння: dK/dt = r(M - K)

    Фізичний сенс:
        Швидкість навчання пропорційна кількості ще не засвоєного матеріалу.
        (M - K) - це матеріал, який студент ще не знає.
        Чим більше вже знає студент, тим менше залишилось вивчити,
        отже швидкість навчання зменшується.

    Параметри:
        t (float): час у днях (параметр обов'язковий для solve_ivp,
                   хоча рівняння автономне)
        K (float): поточний рівень знань студента (%)

    Повертає:
        float: швидкість зміни рівня знань dK/dt (% на день)
    """
    M = 100    # Максимальний рівень знань (100%)
    r = 0.15   # Коефіцієнт швидкості навчання (1/день)

    # Обчислюємо праву частину рівняння
    dK_dt = r * (M - K)

    return dK_dt


# ============================================================================
# 2. РОЗВ'ЯЗАННЯ РІВНЯННЯ ЧИСЕЛЬНО
# ============================================================================

def solve_learning_equation(K0=10, t_span=(0, 30), method='RK45'):
    """
    Розв'язує диференціальне рівняння навчання чисельним методом.

    Параметри:
        K0 (float): початковий рівень знань (%)
        t_span (tuple): інтервал часу (початок, кінець) у днях
        method (str): метод чисельного розв'язання

    Повертає:
        sol: об'єкт розв'язку з методами для отримання значень
    """
    print(f"\n{'='*70}")
    print(f"РОЗВ'ЯЗАННЯ ДИФЕРЕНЦІАЛЬНОГО РІВНЯННЯ")
    print(f"{'='*70}")

    print(f"\nПараметри моделі:")
    print(f"  • Рівняння: dK/dt = 0.15 × (100 - K)")
    print(f"  • Початкова умова: K(0) = {K0}%")
    print(f"  • Період розв'язання: від {t_span$0]} до {t_span$1]} днів")
    print(f"  • Метод: {method} (Runge-Kutta 4-5 порядку)")

    # Розв'язуємо диференціальне рівняння
    sol = solve_ivp(
        fun=learning_rate,           # Функція правої частини рівняння
        t_span=t_span,               # Інтервал часу
        y0=$K0],                     # Початкова умова (має бути масив)
        method=method,               # Метод розв'язання
        dense_output=True,           # Дозволяє інтерполяцію розв'язку
        max_step=0.5                 # Максимальний крок інтегрування
    )

    print(f"\nРівняння успішно розв'язано!")
    print(f"  • Кількість кроків: {sol.t.size}")
    print(f"  • Статус: {'Успіх' if sol.success else 'Помилка'}")

    return sol


def analyze_solution(sol, K0, t_points=None):
    """
    Аналізує розв'язок рівняння в конкретних точках.

    Параметри:
        sol: об'єкт розв'язку
        K0 (float): початковий рівень
        t_points (list): точки часу для аналізу
    """
    if t_points is None:
        t_points = $0, 5, 10, 15, 20, 25, 30]

    print(f"\nАналіз прогресу навчання (K(0) = {K0}%):")
    print(f"{'─'*70}")
    print(f"{'День':<10} {'Рівень знань':<20} {'Засвоєно':<20} {'Залишилось':<20}")
    print(f"{'─'*70}")

    for t in t_points:
        K_t = sol.sol(t)$0]
        learned = K_t - K0
        remaining = 100 - K_t

        # Візуальна шкала прогресу
        progress = int(K_t / 5)
        bar = '█' * progress + '░' * (20 - progress)

        print(f"{t:<10} {K_t:>6.2f}% ${bar}] {learned:>6.2f}% (+) {remaining:>6.2f}% (−)")

    print(f"{'─'*70}")


# ============================================================================
# 3. ДОСЛІДЖЕННЯ ВПЛИВУ ПОЧАТКОВОГО РІВНЯ
# ============================================================================

def find_time_to_target(sol, target=90, max_time=30):
    """
    Знаходить час, коли студент досягає цільового рівня знань.

    Параметри:
        sol: об'єкт розв'язку
        target (float): цільовий рівень знань (%)
        max_time (float): максимальний час пошуку (дні)

    Повертає:
        float: час досягнення цільового рівня або None
    """
    # Створюємо густу сітку точок для пошуку
    t_search = np.linspace(0, max_time, 10000)
    K_search = sol.sol(t_search)$0]

    # Знаходимо перший момент, коли K >= target
    indices = np.where(K_search >= target)$0]

    if len(indices) > 0:
        # Уточнюємо через лінійну інтерполяцію
        idx = indices$0]
        if idx > 0:
            t1, t2 = t_search$idx-1], t_search$idx]
            K1, K2 = K_search$idx-1], K_search$idx]

            # Лінійна інтерполяція
            t_target = t1 + (target - K1) * (t2 - t1) / (K2 - K1)
            return t_target
        else:
            return t_search$idx]
    else:
        return None


def compare_initial_levels():
    """
    Порівнює час досягнення 90% знань для студентів 
    з різним початковим рівнем.
    """
    print(f"\n{'='*70}")
    print(f"ДОСЛІДЖЕННЯ ВПЛИВУ ПОЧАТКОВОГО РІВНЯ ЗНАНЬ")
    print(f"{'='*70}")

    # Початкові рівні для порівняння
    initial_levels = $5, 10, 20]
    target_knowledge = 90

    results = {}

    print(f"\nЦільовий рівень знань: {target_knowledge}%\n")

    # Розв'язуємо рівняння для кожного початкового рівня
    for K0 in initial_levels:
        print(f"\n{'─'*70}")
        print(f"СТУДЕНТ {initial_levels.index(K0) + 1}: K(0) = {K0}%")
        print(f"{'─'*70}")

        # Розв'язуємо рівняння
        sol = solve_learning_equation(K0=K0, t_span=(0, 30))

        # Аналізуємо розв'язок
        analyze_solution(sol, K0, t_points=$0, 10, 20, 30])

        # Знаходимо час досягнення 90%
        time_90 = find_time_to_target(sol, target=target_knowledge, max_time=30)

        # Зберігаємо результати
        results$K0] = {
            'sol': sol,
            'time_90': time_90,
            'K_30': sol.sol(30)$0]
        }

        # Виводимо результат
        if time_90 is not None:
            days = int(time_90)
            hours = int((time_90 - days) * 24)
            print(f"\n Час досягнення {target_knowledge}%: {time_90:.2f} днів ({days} днів {hours} годин)")
        else:
            print(f"\n Рівень {target_knowledge}% не досягнуто за 30 днів")
            print(f"    Рівень на день 30: {results$K0]$'K_30']:.2f}%")

    return results


def print_comparison_table(results):
    """
    Виводить порівняльну таблицю результатів.

    Параметри:
        results (dict): словник з результатами для різних K0
    """
    print(f"\n{'='*70}")
    print(f"ПОРІВНЯЛЬНА ТАБЛИЦЯ РЕЗУЛЬТАТІВ")
    print(f"{'='*70}\n")

    target = 90

    print(f"{'Початковий рівень':<25} {'Час до 90%':<20} {'Рівень на день 30':<25}")
    print(f"{'─'*70}")

    times = $]
    for K0 in sorted(results.keys()):
        time_90 = results$K0]$'time_90']
        K_30 = results$K0]$'K_30']

        if time_90 is not None:
            time_str = f"{time_90:.2f} днів"
            times.append((K0, time_90))
        else:
            time_str = "> 30 днів"

        print(f"K(0) = {K0}% {'':<15} {time_str:<20} {K_30:.2f}%")

    print(f"{'─'*70}")

    # Аналіз різниці в часі
    if len(times) >= 2:
        print(f"\nАНАЛІЗ ВПЛИВУ ПОЧАТКОВОЇ ПІДГОТОВКИ:\n")

        for i in range(len(times) - 1):
            K0_1, t1 = times$i]
            K0_2, t2 = times$i + 1]
            diff = t1 - t2
            percent_faster = (diff / t1) * 100

            print(f"  • Студент з K(0) = {K0_2}% досягає {target}% на {diff:.2f} днів")
            print(f"    ШВИДШЕ, ніж студент з K(0) = {K0_1}%")
            print(f"    (економія часу: {percent_faster:.1f}%)\n")

        # Загальний висновок
        K0_min, t_max = times$0]
        K0_max, t_min = times$-1]
        total_diff = t_max - t_min

        print(f"  ВИСНОВОК:")
        print(f"     Студент з кращою початковою підготовкою (K(0) = {K0_max}%)")
        print(f"     досягає цільового рівня на {total_diff:.2f} днів швидше,")
        print(f"     ніж студент зі слабшою підготовкою (K(0) = {K0_min}%).")
        print(f"\n     Початкова підготовка КРИТИЧНО ВАЖЛИВА для")
        print(f"     швидкості засвоєння матеріалу!")


# ============================================================================
# 4. ВІЗУАЛІЗАЦІЯ РЕЗУЛЬТАТІВ
# ============================================================================

def visualize_learning_curves(results):
    """
    Створює графіки кривих навчання для різних початкових умов.

    Параметри:
        results (dict): словник з результатами
    """
    print(f"\n{'='*70}")
    print(f"ВІЗУАЛІЗАЦІЯ РЕЗУЛЬТАТІВ")
    print(f"{'='*70}")

    # Створюємо фігуру з підграфіками
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('Адаптивна система навчання: Математична модель', 
                 fontsize=16, fontweight='bold')

    # Кольори для різних студентів
    colors = $'#FF6B6B', '#4ECDC4', '#45B7D1']
    labels = $'Студент 1: K(0)=5%', 'Студент 2: K(0)=10%', 'Студент 3: K(0)=20%']

    # ===== ГРАФІК 1: Криві навчання =====
    ax1 = axes$0, 0]

    t_plot = np.linspace(0, 30, 500)

    for i, K0 in enumerate(sorted(results.keys())):
        sol = results$K0]$'sol']
        K_plot = sol.sol(t_plot)$0]

        ax1.plot(t_plot, K_plot, color=colors$i], linewidth=2.5, 
                label=labels$i], marker='o', markevery=50, markersize=5)

        # Позначаємо точку досягнення 90%
        time_90 = results$K0]$'time_90']
        if time_90 is not None:
            ax1.plot(time_90, 90, 'X', color=colors$i], markersize=12,
                    markeredgewidth=2, markeredgecolor='black')

    # Горизонтальна лінія на 90%
    ax1.axhline(y=90, color='orange', linestyle='--', linewidth=2,
               label='Цільовий рівень (90%)', alpha=0.7)
    ax1.axhline(y=100, color='gray', linestyle=':', linewidth=1, alpha=0.5)

    ax1.set_xlabel('Час (дні)', fontsize=12, fontweight='bold')
    ax1.set_ylabel('Рівень знань K(t) (%)', fontsize=12, fontweight='bold')
    ax1.set_title('Криві навчання для різних початкових рівнів', 
                 fontsize=13, fontweight='bold')
    ax1.grid(True, alpha=0.3, linestyle='--')
    ax1.legend(fontsize=10, loc='lower right')
    ax1.set_ylim($0, 105])
    ax1.set_xlim($0, 30])

    # ===== ГРАФІК 2: Швидкість навчання =====
    ax2 = axes$0, 1]

    for i, K0 in enumerate(sorted(results.keys())):
        sol = results$K0]$'sol']
        K_plot = sol.sol(t_plot)$0]

        # Обчислюємо швидкість dK/dt
        dK_dt = np.array($learning_rate(t, K) for t, K in zip(t_plot, K_plot)])

        ax2.plot(t_plot, dK_dt, color=colors$i], linewidth=2.5,
                label=labels$i], marker='s', markevery=50, markersize=5)

    ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.5, alpha=0.5)
    ax2.set_xlabel('Час (дні)', fontsize=12, fontweight='bold')
    ax2.set_ylabel('Швидкість навчання dK/dt (% на день)', fontsize=12, fontweight='bold')
    ax2.set_title('Швидкість засвоєння матеріалу', fontsize=13, fontweight='bold')
    ax2.grid(True, alpha=0.3, linestyle='--')
    ax2.legend(fontsize=10)
    ax2.set_xlim($0, 30])

    # ===== ГРАФІК 3: Залишок матеріалу (логарифм) =====
    ax3 = axes$1, 0]

    for i, K0 in enumerate(sorted(results.keys())):
        sol = results$K0]$'sol']
        K_plot = sol.sol(t_plot)$0]
        remaining = 100 - K_plot

        ax3.semilogy(t_plot, remaining, color=colors$i], linewidth=2.5,
                    label=labels$i], marker='^', markevery=50, markersize=5)

    ax3.axhline(y=10, color='orange', linestyle='--', linewidth=2,
               label='10% залишилось', alpha=0.7)
    ax3.set_xlabel('Час (дні)', fontsize=12, fontweight='bold')
    ax3.set_ylabel('Залишок матеріалу (100 - K) %', fontsize=12, fontweight='bold')
    ax3.set_title('Експоненціальне спадання незасвоєного матеріалу', 
                 fontsize=13, fontweight='bold')
    ax3.grid(True, alpha=0.3, linestyle='--', which='both')
    ax3.legend(fontsize=10)
    ax3.set_xlim($0, 30])

    # ===== ГРАФІК 4: Порівняльна діаграма часу =====
    ax4 = axes$1, 1]

    K0_values = $]
    times_90 = $]

    for K0 in sorted(results.keys()):
        time_90 = results$K0]$'time_90']
        if time_90 is not None:
            K0_values.append(f'K(0)={K0}%')
            times_90.append(time_90)

    bars = ax4.barh(K0_values, times_90, color=colors$:len(times_90)], 
                    edgecolor='black', linewidth=1.5, alpha=0.8)

    # Додаємо значення на стовпчики
    for i, (bar, time) in enumerate(zip(bars, times_90)):
        ax4.text(time + 0.5, i, f'{time:.2f} днів', 
                va='center', fontsize=11, fontweight='bold')

    ax4.set_xlabel('Час досягнення 90% знань (дні)', fontsize=12, fontweight='bold')
    ax4.set_ylabel('Початковий рівень', fontsize=12, fontweight='bold')
    ax4.set_title('Порівняння часу навчання', fontsize=13, fontweight='bold')
    ax4.grid(True, alpha=0.3, linestyle='--', axis='x')
    ax4.set_xlim($0, max(times_90) * 1.15])

    plt.tight_layout()
    plt.show()

    print("\nГрафіки успішно побудовано!")


# ============================================================================
# ГОЛОВНА ПРОГРАМА
# ============================================================================

def main():
    """
    Головна функція для виконання всіх кроків завдання.
    """
    print("\n" + "="*70)
    print("  АДАПТИВНА СИСТЕМА НАВЧАННЯ - МАТЕМАТИЧНА МОДЕЛЬ")
    print("  Моделювання процесу засвоєння знань студентами")
    print("="*70)

    # Крок 1: Реалізація моделі
    print(f"\n{'='*70}")
    print("1. РЕАЛІЗАЦІЯ МОДЕЛІ")
    print(f"{'='*70}")

    print("\nДиференціальне рівняння навчання:")
    print("  ┌────────────────────────────┐")
    print("  │  dK/dt = r(M - K)          │")
    print("  │  dK/dt = 0.15(100 - K)     │")
    print("  └────────────────────────────┘")

    print("\nДе:")
    print("  • K(t) - рівень знань студента в момент t (%)")
    print("  • M = 100 - максимальний рівень знань (%)")
    print("  • r = 0.15 - коефіцієнт швидкості навчання (1/день)")
    print("  • (M - K) - кількість ще не засвоєного матеріалу")

    print("\nФункція learning_rate(t, K) успішно реалізована!")

    # Крок 2: Розв'язання рівняння
    print(f"\n{'='*70}")
    print("2. ЧИСЕЛЬНЕ РОЗВ'ЯЗАННЯ РІВНЯННЯ")
    print(f"{'='*70}")

    sol_main = solve_learning_equation(K0=10, t_span=(0, 30))
    analyze_solution(sol_main, K0=10)

    # Крок 3: Дослідження впливу початкового рівня
    results = compare_initial_levels()
    print_comparison_table(results)

    # Візуалізація
    print("\n" + "─"*70)
    input("Натисніть Enter для побудови графіків...")
    visualize_learning_curves(results)

    # Завершення
    print("\n" + "="*70)
    print("ПРОГРАМА УСПІШНО ЗАВЕРШЕНА")
    print("="*70)
    print("\nРезультати аналізу показують важливість початкової підготовки")
    print("для ефективного засвоєння нового матеріалу в EdTech-платформах.")
    print("="*70 + "\n")


if __name__ == "__main__":
    main()
```

## Розв'язок
### 1. Аналітичний інтеграл за перші 7 днів

Маємо інтенсивність реєстрацій: $ f(t) = 500 e^{-0.3t}. $

Потрібно обчислити: $ I_7 = \int_0^7 500 e^{-0.3t},dt. $

Первісна: $ \int 500 e^{-0.3t},dt = 500 \cdot \int e^{-0.3t},dt = 500 \cdot \left(-\frac{1}{0.3} e^{-0.3t}\right) = -\frac{500}{0.3} e^{-0.3t} = -\frac{5000}{3} e^{-0.3t}. $

За формулою Ньютона–Лейбніца: $ I_7 = \left[-\frac{5000}{3} e^{-0.3t}\right]_{0}^{7} = -\frac{5000}{3} e^{-0.3\cdot 7} + \frac{5000}{3} e^{0} = \frac{5000}{3}\left(1 - e^{-2.1}\right). $

Чисельно: $ e^{-2.1} \approx 0.122456,\quad 1 - e^{-2.1} \approx 0.877544, $ $ I_7 \approx \frac{5000}{3} \cdot 0.877544 \approx 1462.6. $

Аналітичний результат: $ \boxed{I_7 = \frac{5000}{3}\left(1 - e^{-2.1}\right) \approx 1463 \text{ реєстрацій}.} $

### 2. Чисельна перевірка інтегралу (Python + quad)

```python
from scipy.integrate import quad

def f(t):
    return 500 * np.exp(-0.3 * t)

# інтеграл від 0 до 7
I7_num, I7_err = quad(f, 0, 7)

print("Аналітичний I7 ≈", (5000/3) * (1 - np.exp(-2.1)))
print("Чисельний I7   ≈", I7_num)
print("Оцінка похибки  ≈", I7_err)
```

### 3. Теоретичний максимум реєстрацій (невласний інтеграл)

Потрібно: $ I_{\infty} = \int_0^{\infty} 500 e^{-0.3t},dt. $

Використовуємо ту саму первісну: $ F(t) = -\frac{5000}{3} e^{-0.3t}. $

Тоді: $ I_{\infty} = \lim_{b \to \infty} \int_0^{b} 500 e^{-0.3t},dt = \lim_{b \to \infty} \left[-\frac{5000}{3} e^{-0.3t}\right]{0}^{b} = \lim_{b \to \infty} \left(-\frac{5000}{3} e^{-0.3b} + \frac{5000}{3}\right). $

Оскільки: $ \lim_{b \to \infty} e^{-0.3b} = 0, $ маємо: $ I_{\infty} = \frac{5000}{3} \approx 1666.7. $

Теоретичний максимум: $ \boxed{I_{\infty} = \frac{5000}{3} \approx 1667 \text{ реєстрацій}.} $

```python
I_inf_num, I_inf_err = quad(f, 0, np.inf)
print("Аналітичний I_inf ≈", 5000/3)
print("Чисельний I_inf   ≈", I_inf_num)
print("Оцінка похибки    ≈", I_inf_err)
```

## Розв'язок
### Аналітичні частинні похідні

Маємо функцію $ f(x, y) = 0.5x^2 + 0.3y^2 + 0.2xy + 10x + 5y. $

Частинна похідна за (x): $ \frac{\partial f}{\partial x} = 0.5 \cdot 2x + 0.2y + 10 = x + 0.2y + 10. $

Частинна похідна за (y): $ \frac{\partial f}{\partial y} = 0.3 \cdot 2y + 0.2x + 5 = 0.6y + 0.2x + 5. $

$ \boxed{ \frac{\partial f}{\partial x} = x + 0.2y + 10,\quad \frac{\partial f}{\partial y} = 0.6y + 0.2x + 5. } $


### Чисельний градієнт у точці $(10,20)$ (Python)

```python
from scipy.optimize import approx_fprime

def f_vec(v):
    x, y = v
    return 0.5*x**2 + 0.3*y**2 + 0.2*x*y + 10*x + 5*y

x0 = np.array([10.0, 20.0])
eps = np.sqrt(np.finfo(float).eps)

grad_num = approx_fprime(x0, f_vec, eps)
print("Чисельний градієнт у (10,20):", grad_num)
```

Очікуваний чисельний результат: $ \nabla f(10,20) \approx (24,\ 19). $

### Порівняння з аналітичними похідними

Аналітично в точці ((10,20)):

$ \frac{\partial f}{\partial x}(10,20) = 10 + 0.2\cdot 20 + 10 = 10 + 4 + 10 = 24, $

$ \frac{\partial f}{\partial y}(10,20) = 0.6\cdot 20 + 0.2\cdot 10 + 5 = 12 + 2 + 5 = 19. $

$ \boxed{\nabla f(10,20) = (24,\ 19).} $

Чисельний градієнт з approx_fprime збігається з аналітичним (з точністю до машинної похибки).

### Лінійна апроксимація зміни функції

Формула лінійної апроксимації: $ \Delta f \approx \frac{\partial f}{\partial x}(x_0,y_0)\cdot \Delta x

\frac{\partial f}{\partial y}(x_0,y_0)\cdot \Delta y. $

Нехай $(x_0,y_0) = (10,20)$,$\Delta x = +0.5$, $\Delta y = -0.3$.

Тоді: $ \Delta f_{\text{approx}} \approx 24 \cdot 0.5 + 19 \cdot (-0.3) = 12 - 5.7 = 6.3. $

### Точна зміна значення функції

Обчислимо: $ f(10,20) = 0.5\cdot 10^2 + 0.3\cdot 20^2 + 0.2\cdot 10\cdot 20 + 10\cdot 10 + 5\cdot 20. $

$ f(10,20) = 0.5\cdot 100 + 0.3\cdot 400 + 0.2\cdot 200 + 100 + 100 = 50 + 120 + 40 + 100 + 100 = 410. $

Для $(10.5, 19.7)$:

$ f(10.5,19.7) = 0.5\cdot 10.5^2 + 0.3\cdot 19.7^2 + 0.2\cdot 10.5\cdot 19.7 + 10\cdot 10.5 + 5\cdot 19.7. $

Чисельно: $ f(10.5,19.7) \approx 416.422, $ $ \Delta f_{\text{exact}} = f(10.5,19.7) - f(10,20) \approx 416.422 - 410 = 6.422. $

Отже: $ \boxed{ \Delta f_{\text{approx}} \approx 6.3,\quad \Delta f_{\text{exact}} \approx 6.422. } $

Лінійна апроксимація дає дуже близьке значення до точної зміни, що добре ілюструє роботу градієнта як локальної лінійної моделі функції.
