📐Нагадуємо, чим менший кут між цими векторами у багатовимірному просторі, тим більше схожі об'єкти. Мірою цієї схожості є **косинусна подібність**. Вона показує, наскільки вектори "дивляться" в один бік. Значення коливається від -1 (протилежні) до 1 (ідентичні).

Формула косинусної подібності:
$\cos \theta = \cfrac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \cdot \|\mathbf{v}\|}$,

де $\mathbf{u} \cdot \mathbf{v}$ — скалярний добуток, а $\|\mathbf{u}\|, \|\mathbf{v}\|$ — довжина (норма) вектора.

Рішення:

## 1. Аналітична частина

### 1.1. Норма вектора користувача

Вектор користувача:

$u = (8, 2, 5)$

Обчислюємо норму:

$‖u‖ = \sqrt{8^2 + 2^2 + 5^2}$  
$‖u‖ = \sqrt{64 + 4 + 25}$  
$‖u‖ = \sqrt{93}$

---

### 1.2. Косинусна подібність з фільмами

Формула косинусної подібності:

$cos(u, v) = \frac{(u · v)}{‖u‖ · ‖v‖}$

---

#### Фільм A

$v_A = (9, 1, 2)$

$u · v_A = 8·9 + 2·1 + 5·2 = 84$  
$‖v_A‖ = \sqrt{9^2 + 1^2 + 2^2} = \sqrt{86}$

$cos(u, v_A) = \frac{84}{\sqrt{93} · \sqrt{86}}$

---

#### Фільм B

$v_B = (1, 9, 8)$

$u · v_B = 66$ 
$‖v_B‖ = \sqrt{146}$

$cos(u, v_B) = \frac{66}{\sqrt{93} · \sqrt{146}}$

---

#### Фільм C

$v_C = (7, 2, 6)$

$u · v_C = 90$  
$‖v_C‖ = \sqrt{89}$

$cos(u, v_C) = \frac{90}{\sqrt{93} · \sqrt{89}}$

---

### 1.3. Висновок

Найбільше значення косинусної подібності має фільм **C**, тому він найкраще відповідає профілю користувача.


## 2. Реалізація в Python

````python
import numpy as np

# Вектори
u = np.array([8, 2, 5])
vA = np.array([9, 1, 2])
vB = np.array([1, 9, 8])
vC = np.array([7, 2, 6])

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# Обчислення
sim_A = cosine_similarity(u, vA)
sim_B = cosine_similarity(u, vB)
sim_C = cosine_similarity(u, vC)

print("||u|| =", np.linalg.norm(u))
print("cos(u, A) =", sim_A)
print("cos(u, B) =", sim_B)
print("cos(u, C) =", sim_C)

# Вибір найкращого фільму
sims = {
    "Film A": sim_A,
    "Film B": sim_B,
    "Film C": sim_C
}

best_film = max(sims, key=sims.get)

print("\nBest match:", best_film, "with similarity =", sims[best_film])
````

Рішення:

Є три моделі дронів:
- x1 — "Розвідник"
- x2 — "Камікадзе"
- x3 — "Вантажний"

Матриця витрат деталей на один дрон $A$:

Розвідник: 4 мотори, 1 контролер, 1 акумулятор  
Камікадзе: 4 мотори, 1 контролер, 2 акумулятори  
Вантажний: 6 моторів, 2 контролери, 4 акумулятори  

Матриця $A$ (рядки — типи деталей, стовпці — типи дронів):

$
\begin{pmatrix} 
  4 & 4 & 6 \\  
  1 & 1 & 2 \\  
  1 & 2 & 4  
\end{pmatrix}
$

Вектор витрат за зміну $b$:

$b = (460, 130, 240)$

Потрібно знайти $x = (x_1, x_2, x_3)$, де:
$A$ * x = $b$

---

#### Кроки розв'язання

1. Створити матрицю $A$ та вектор $b$.  
2. Перевірити, чи існує єдиний розв'язок: det($A$) ≠ 0.  
3. Розв'язати систему: `x = np.linalg.solve(A, b)`.  
4. Виконати перевірку: $A$ @ x має дорівнювати $b$.  
5. Вивести кількість дронів кожного типу.


````python
import numpy as np

# 1. Матриця коефіцієнтів A
A = np.array([
    [4, 4, 6],   # мотори
    [1, 1, 2],   # контролери
    [1, 2, 4]    # акумулятори
])

# 2. Вектор витрат b
b = np.array([460, 130, 240])

# Перевірка визначника
detA = np.linalg.det(A)
print("Детермінант A:", detA)

if abs(detA) < 1e-9:
    print("Система не має єдиного розв'язку.")
else:
    # 3. Розв'язання системи
    x = np.linalg.solve(A, b)
    print("\nКількість зібраних дронів:")
    print("Розвідник (x1):", x[0])
    print("Камікадзе (x2):", x[1])
    print("Вантажний (x3):", x[2])

    # 4. Перевірка
    check = A @ x
    print("\nПеревірка A * x:")
    print(check)
    print("\nОчікуваний вектор b:")
    print(b)
````

1. Сформувати матрицю $A$:

   A має містити:
   - перший стовпчик — значення $t$
   - другий стовпчик — одиниці (для вільного члена $b$)

   Тобто:

$A =
\begin{pmatrix} 
     1 & 1 \\
     2 & 1 \\
     3 & 1 \\
     4 & 1 \\
     5 & 1
\end{pmatrix} 
$

2. Знайти коефіцієнти k і b за допомогою:

```
x = [k, b]
x = np.linalg.lstsq(A, y, rcond=None)[0]
```

3. Обчислити прогноз на 6-ту годину:

   `y_hat_6 = k * 6 + b`

4. Вивести рівняння тренду та прогноз.

## Опціонально: нормальне рівняння

Метод МНК можна записати як:

$(A^T A) x = A^T y$

Кроки:

1. Обчислити $M = A^T @ A$
2. Обчислити $v = A^T @ y$
3. Розв'язати систему:

   `x_manual = np.linalg.solve(M, v)`

4. Порівняти `x_manual` з результатом з `lstsq` — вони мають збігатися.


````python
import numpy as np

# Дані
t = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([22, 28, 37, 45, 53], dtype=float)

# 1. Формуємо матрицю A: стовпчик t і стовпчик одиниць
A = np.column_stack((t, np.ones_like(t)))

print("Matrix A:")
print(A)
print("\nVector y:")
print(y)

# 2. Розв'язання через np.linalg.lstsq
coeffs, residuals, rank, s = np.linalg.lstsq(A, y, rcond=None)
k, b = coeffs

print("\nКоефіцієнти (через lstsq):")
print("k (нахил) =", k)
print("b (зсув)  =", b)

# 3. Прогноз на 6-ту годину
t_future = 6
y_hat_6 = k * t_future + b
print(f"\nПрогнозоване навантаження на {t_future}-ту годину: y = {y_hat_6}")

print(f"\nРівняння тренду: y = {k} * t + {b}")

# 4*. Реалізація через нормальне рівняння
AT = A.T
ATA = AT @ A
ATy = AT @ y

x_manual = np.linalg.solve(ATA, ATy)
k_manual, b_manual = x_manual

print("\nКоефіцієнти (через нормальне рівняння):")
print("k_manual =", k_manual)
print("b_manual =", b_manual)

# Порівняння
print("\nПорівняння результатів:")
print("Різниця k:", k - k_manual)
print("Різниця b:", b - b_manual)
````
