# Задача 1. Аналітика помилок у логах

### Умова

Ви працюєте DevOps-інженером і аналізуєте логи системи моніторингу. За добу зафіксовано 500 подій. З них:

- 85 — помилки тільки бази даних з тегом `DB`
- 60 — помилки  тільки мережі з тегом `NET`
- 35 — помилки, що мають обидва теги одночасно `DB` та `NET`

Позначимо:

- $\Omega$ — множина всіх 500 подій
- $A$ — подія "помилка бази даних" з тегом `DB`
- $B$ — подія "помилка мережі" з тегом `NET`

**Завдання:**

1. Яка ймовірність, що випадково вибрана подія — помилка DB?
2. Яка ймовірність, що подія має тег DB **або** NET?
3. Яка ймовірність, що подія має тег DB, але **не** NET?

Критерії оцінювання (25 балів)

| **Критерій** | **Бали** | **Вимоги** |
| --- | --- | --- |
| **Класична ймовірність** | 7 | Правильно визначено простір подій та обчислено $P(A)$. |
| **Об'єднання подій** | 10 | Правильно обчислено ймовірність події "DB або NET", враховано перетин $P(A \cup B)$. |
| **Різниця подій** | 8 | Правильно обчислено ймовірність події "DB, але не NET" $P(A \setminus B)$. |

### Ймовірність помилки DB

Кількість подій з тегом DB:

$ |A| = 85 + 35 = 120 $

(85 тільки DB + 35 з DB і NET).

Тоді класична ймовірність:

$ P(A) = \frac{|A|}{|\Omega|} = \frac{120}{500} = 0.24. $

Ймовірність, що подія має тег DB або NET

Кількість подій з тегом NET:

$ |B| = 60 + 35 = 95 $

(60 тільки NET + 35 з DB і NET).

Використовуємо формулу:

$ P(A \cup B) = \frac{|A| + |B| - |A \cap B|}{|\Omega|} $

$ |A \cup B| = 120 + 95 - 35 = 180 $

$ P(A \cup B) = \frac{180}{500} = 0.36. $

Ймовірність, що подія має тег DB, але не NET

Це події, які належать (A), але не належать (B):

$ |A \setminus B| = 85 $

(тільки DB).

$ P(A \setminus B) = \frac{85}{500} = 0.17. $

Відповіді:

$P(\text{DB}) = 0.24$

$P(\text{DB або NET}) = 0.36$

$P(\text{DB, але не NET}) = 0.17$

### 1. Ймовірність, що всі 3 батареї справні (правило множення)

На складі:

8 справних,

4 дефектні,усього 12.

Вибираємо 3 без повернення.

Ймовірність, що перша батарея справна: $ P_1 = \frac{8}{12}. $

Після цього лишається 7 справних із 11: $ P_2 = \frac{7}{11}. $

Після двох справних лишається 6 справних із 10: $ P_3 = \frac{6}{10}. $

Ймовірність, що всі три справні: $ P(\text{3 справні}) = P_1 \cdot P_2 \cdot P_3 = \frac{8}{12} \cdot \frac{7}{11} \cdot \frac{6}{10} = \frac{8 \cdot 7 \cdot 6}{12 \cdot 11 \cdot 10} = \frac{336}{1320} = \frac{14}{55} \approx 0.2545. $

### 2. Та сама ймовірність через комбінаторику

Кількість способів вибрати 3 батареї з 12: $ C_{12}^{3} = \frac{12!}{3! \cdot 9!} = 220. $

Кількість способів вибрати 3 справні з 8: $ C_{8}^{3} = \frac{8!}{3! \cdot 5!} = 56. $

Ймовірність: $ P(\text{3 справні}) = \frac{C_{8}^{3}}{C_{12}^{3}} = \frac{56}{220} = \frac{14}{55} \approx 0.2545. $

Результат збігається з пунктом 1.

### 3. Ймовірність, що рівно 2 батареї справні

#### 3.1. Через правило множення (усі порядки)

Рівно 2 справні й 1 дефектна. Можливі порядки:

С, С, Д

С, Д, С

Д, С, С

Позначимо С — справна, Д — дефектна.

Варіант 1: С, С, Д

$ P(\text{С, С, Д}) = \frac{8}{12} \cdot \frac{7}{11} \cdot \frac{4}{10} = \frac{8 \cdot 7 \cdot 4}{12 \cdot 11 \cdot 10} = \frac{224}{1320}. $

Варіант 2: С, Д, С

$ P(\text{С, Д, С}) = \frac{8}{12} \cdot \frac{4}{11} \cdot \frac{7}{10} = \frac{8 \cdot 4 \cdot 7}{12 \cdot 11 \cdot 10} = \frac{224}{1320}. $

Варіант 3: Д, С, С

$ P(\text{Д, С, С}) = \frac{4}{12} \cdot \frac{8}{11} \cdot \frac{7}{10} = \frac{4 \cdot 8 \cdot 7}{12 \cdot 11 \cdot 10} = \frac{224}{1320}. $

Сумарно: $ P(\text{рівно 2 справні}) = 3 \cdot \frac{224}{1320} = \frac{672}{1320} = \frac{56}{110} = \frac{28}{55} \approx 0.5091. $

#### 3.2. Через комбінаторику

Рівно 2 справні й 1 дефектна:

вибрати 2 справні з 8: $ C_{8}^{2} = \frac{8!}{2! \cdot 6!} = 28, $

вибрати 1 дефектну з 4: $ C_{4}^{1} = 4. $

Кількість сприятливих наборів: $ N_{\text{good}} = C_{8}^{2} \cdot C_{4}^{1} = 28 \cdot 4 = 112. $

Загальна кількість наборів: $ C_{12}^{3} = 220. $

Ймовірність: $ P(\text{рівно 2 справні}) = \frac{112}{220} = \frac{56}{110} = \frac{28}{55} \approx 0.5091. $

Результат збігається з методом правила множення.

### Підсумок:

$P(\text{3 справні}) = \dfrac{14}{55} \approx 0.2545$.

$P(\text{рівно 2 справні}) = \dfrac{28}{55} \approx 0.5091$.

# Задача 3. Надійність розподіленої системи

### Умова

Ви проєктуєте розподілену систему, яка складається з 4 незалежних вузлів. Ймовірність відмови кожного вузла протягом однієї години становить:

- Вузол A: 0.02
- Вузол B: 0.05
- Вузол C: 0.03
- Вузол D: 0.04

Позначимо:

- $A_1$ — подія "вузол A відмовив протягом години"
- $A_2$ — подія "вузол B відмовив протягом години"
- $A_3$ — подія "вузол C відмовив протягом години"
- $A_4$ — подія "вузол D відмовив протягом години"

Завдання

1. Яка ймовірність, що протягом години всі 4 вузли працюють?
2. Яка ймовірність, що протягом години хоча б один вузол відмовить?
3. Яка ймовірність, що протягом години відмовлять рівно два вузли?

Критерії оцінювання (25 балів)

| **Критерій** | **Бали** | **Вимоги** |
| --- | --- | --- |
| **Всі працюють** | 7 | Правильно застосовано **незалежність подій** та обчислено добуток ймовірностей роботи кожного вузла. |
| **Хоча б один відмовить** | 8 | Застосовано формулу через **протилежну подію**, результат узгоджується з пунктом 1. |
| **Рівно два відмовлять** | 10 | Перераховано всі 6 несумісних комбінацій пар вузлів, правильно обчислено кожну ймовірність (через незалежність) та підсумовано. |

### 1. Ймовірність, що всі 4 вузли працюють протягом години

Ймовірність відмови вузла:

$P(A_1) = 0.02$

$P(A_2) = 0.05$

$P(A_3) = 0.03$

$P(A_4) = 0.04$

Тоді ймовірність, що вузол працює:

$P(\overline{A_1}) = 1 - 0.02 = 0.98$

$P(\overline{A_2}) = 1 - 0.05 = 0.95$

$P(\overline{A_3}) = 1 - 0.03 = 0.97$

$P(\overline{A_4}) = 1 - 0.04 = 0.96$

Оскільки вузли незалежні:

$ P(\text{всі працюють}) = 0.98 \cdot 0.95 \cdot 0.97 \cdot 0.96 \approx 0.872. $

### 2. Ймовірність, що хоча б один вузол відмовить

Подія «хоча б один відмовить» — це протилежна до «всі працюють»:

$ P(\text{хоча б один відмовить}) = 1 - P(\text{всі працюють}) = 1 - 0.872 \approx 0.128. $

### 3. Ймовірність, що відмовлять рівно два вузли

Потрібно врахувати всі пари вузлів, які відмовляють, а решта працюють. Вузли незалежні, тому ймовірність кожної конкретної комбінації — добуток відповідних ймовірностей.

Позначимо:

$p_A = 0.02,\ p_B = 0.05,\ p_C = 0.03,\ p_D = 0.04$

$q_A = 0.98,\ q_B = 0.95,\ q_C = 0.97,\ q_D = 0.96$

Маємо 6 пар:

Відмовляють A і B, C і D працюють: $ P_{AB} = p_A \cdot p_B \cdot q_C \cdot q_D = 0.02 \cdot 0.05 \cdot 0.97 \cdot 0.96 \approx 0.0009312. $

Відмовляють A і C, B і D працюють: $ P_{AC} = p_A \cdot p_C \cdot q_B \cdot q_D = 0.02 \cdot 0.03 \cdot 0.95 \cdot 0.96 \approx 0.0005472. $

Відмовляють A і D, B і C працюють: $ P_{AD} = p_A \cdot p_D \cdot q_B \cdot q_C = 0.02 \cdot 0.04 \cdot 0.95 \cdot 0.97 \approx 0.0007372. $

Відмовляють B і C, A і D працюють: $ P_{BC} = p_B \cdot p_C \cdot q_A \cdot q_D = 0.05 \cdot 0.03 \cdot 0.98 \cdot 0.96 \approx 0.0014112. $

Відмовляють B і D, A і C працюють: $ P_{BD} = p_B \cdot p_D \cdot q_A \cdot q_C = 0.05 \cdot 0.04 \cdot 0.98 \cdot 0.97 \approx 0.0019028. $

Відмовляють C і D, A і B працюють: $ P_{CD} = p_C \cdot p_D \cdot q_A \cdot q_B = 0.03 \cdot 0.04 \cdot 0.98 \cdot 0.95 \approx 0.0011172. $

Сумарно:

$ P(\text{рівно два вузли відмовлять}) = P_{AB} + P_{AC} + P_{AD} + P_{BC} + P_{BD} + P_{CD} \approx 0.00665. $

### Підсумок:

$P(\text{всі працюють}) \approx 0.872$

$P(\text{хоча б один відмовить}) \approx 0.128$

$P(\text{рівно два відмовлять}) \approx 0.00665$

# Задача 4. Визначення джерела трафіку

### Умова

Інтернет-магазин отримує трафік з трьох джерел:

- Пошукова реклама: 50% трафіку
- Соцмережі: 30% трафіку
- Email-розсилка: 20% трафіку

Конверсія (ймовірність покупки) для кожного джерела:

- Пошукова реклама: 4%
- Соцмережі: 2%
- Email-розсилка: 8%

Позначимо:

- $H_1$ — подія "відвідувач прийшов з пошукової реклами"
- $H_2$  — подія "відвідувач прийшов з соцмереж"
- $H_3$ — подія "відвідувач прийшов з email-розсилки"
- $A$ — подія "відвідувач зробив покупку"

Завдання

1. Яка загальна ймовірність, що випадковий відвідувач зробить покупку?
2. Відвідувач зробив покупку. Яка ймовірність, що він прийшов з email-розсилки?
3. Напишіть функцію `bayes_source(prior_probs, conversion_rates)`, яка для довільної кількості джерел обчислює апостеріорні ймовірності кожного джерела після здійснення покупки.
Параметри функції:
    - `prior_probs` — список апріорних ймовірностей кожного джерела
    - `conversion_rates` — список конверсій для кожного джерела
    
    Функція повертає список апостеріорних ймовірностей для кожного джерела.

### Критерії оцінювання (25 балів)

| **Критерій** | **Бали** | **Вимоги** |
| --- | --- | --- |
| **Повна ймовірність** | 7 | Правильно застосовано формулу повної ймовірності для обчислення загальної ймовірності покупки. |
| **Теорема Баєса** | 8 | Правильно застосовано теорему Баєса для обчислення апостеріорної ймовірності  |
| **Програмна реалізація** | 10 | Функція `bayes_source` коректно працює для довільної кількості джерел, повертає список усіх апостеріорних ймовірностей. |

### 1. Загальна ймовірність покупки (формула повної ймовірності)

Джерела трафіку:

$ P(H_1)=0.5,\quad P(H_2)=0.3,\quad P(H_3)=0.2 $

Конверсії:

$ P(A|H_1)=0.04,\quad P(A|H_2)=0.02,\quad P(A|H_3)=0.08 $

Формула повної ймовірності:

$ P(A)=\sum_{i=1}^{3} P(H_i)\cdot P(A|H_i) $

Обчислюємо:

$ P(A)=0.5\cdot 0.04 + 0.3\cdot 0.02 + 0.2\cdot 0.08 $

$ P(A)=0.02 + 0.006 + 0.016 = 0.042 $

Загальна ймовірність покупки:

$ \boxed{P(A)=0.042\quad (4.2\%)} $

### 2. Ймовірність, що покупець прийшов з email-розсилки (теорема Баєса)

Потрібно знайти:

$ P(H_3|A)=\frac{P(H_3)\cdot P(A|H_3)}{P(A)} $

Підставляємо:

$ P(H_3|A)=\frac{0.2\cdot 0.08}{0.042} $

$ P(H_3|A)=\frac{0.016}{0.042}\approx 0.381 $

Ймовірність:

$ \boxed{P(H_3|A)\approx 0.381\quad (38.1\%)} $

### 3. Функція bayes_source(prior_probs, conversion_rates)

Вона має працювати для будь-якої кількості джерел.

Формула:

$ P(H_i|A)=\frac{P(H_i)\cdot P(A|H_i)}{\sum_j P(H_j)\cdot P(A|H_j)} $

```python
def bayes_source(prior_probs, conversion_rates):
    """
    Обчислює апостеріорні ймовірності джерел трафіку після покупки.
    
    prior_probs        — список апріорних ймовірностей P(H_i)
    conversion_rates   — список конверсій P(A|H_i)
    """
    
    # знаменник формули Баєса — повна ймовірність покупки
    total_prob = sum(p * c for p, c in zip(prior_probs, conversion_rates))
    
    # апостеріорні ймовірності
    posterior = [(p * c) / total_prob for p, c in zip(prior_probs, conversion_rates)]
    
    return posterior


# Дані задачі
prior = [0.5, 0.3, 0.2]
conversion = [0.04, 0.02, 0.08]

# Виклик функції
post = bayes_source(prior, conversion)

print("Апостеріорні ймовірності джерел після покупки:")
for i, p in enumerate(post, start=1):
    print(f"H{i}: {p:.4f}")
```
